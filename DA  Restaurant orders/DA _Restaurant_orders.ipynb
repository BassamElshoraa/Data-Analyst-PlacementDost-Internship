Display the item_name and the price difference from the previous and next menu item

```python
%%sql

SELECT 
    item_name,
    price,
    price - LAG (price) OVER (ORDER BY price) AS price_difference_previous,
    LEAD (price) OVER (ORDER BY price) - price AS price_difference_next
FROM 
    menu_items
LIMIT 20
```

Unpivot the menu_items table to show a list of menu item properties (item_id, item_name, category, price).

```python
%%sql

SELECT 
    menu_item_id AS item_id,
    item_name,
    'category' AS property,
    category AS value
FROM 
    menu_items
UNION ALL
SELECT 
    menu_item_id AS item_id,
    item_name,
    'price' AS property,
    CAST(price AS CHAR) AS value
FROM 
    menu_items
```

Write a dynamic SQL query that allows users to filter menu items based on category and price range.

```python
%%sql

SET @CategoryFilter = NULL;
SET @MinPrice = 9;
SET @MaxPrice = 15;

SET @SQLQuery = 'SELECT menu_item_id, item_name, category, price FROM menu_items WHERE 1 = 1'; -- 1=1 for easy appending of conditions

IF @CategoryFilter IS NOT NULL THEN
    SET @SQLQuery = CONCAT(@SQLQuery, ' AND category = ', @CategoryFilter);
END IF;

IF @MinPrice IS NOT NULL AND @MaxPrice IS NOT NULL THEN
    SET @SQLQuery = CONCAT(@SQLQuery, ' AND price BETWEEN ', @MinPrice, ' AND ', @MaxPrice);
END IF;

EXECUTE IMMEDIATE @SQLQuery;
```

Create a stored procedure that takes a menu category as input and returns the average price for that category.

```python
%%sql

CREATE PROCEDURE GetAveragePriceByCategory(IN category_name VARCHAR(255), OUT avg_price DECIMAL(10, 2))
BEGIN
    -- Declare a variable to store the calculated average price
    DECLARE avg_price_temp DECIMAL(10, 2);
    
    -- Calculate the average price for the specified category
    SELECT AVG(price) INTO avg_price_temp
    FROM menu_items
    WHERE category = category_name;
    
    -- Set the output parameter with the calculated average price
    SET avg_price = avg_price_temp;
END;
```

Design a trigger that updates a log table whenever a new order is inserted into the order_details table.

```python
%%sql

#Create the order_log table
CREATE TABLE IF NOT EXISTS order_log (
    log_id INT AUTO_INCREMENT PRIMARY KEY,
    order_details_id INT,
    order_id INT,
    order_date DATE,
    order_time TIME,
    item_id INT
);
```

```python
%%sql

#Create the trigger
DROP TRIGGER IF EXISTS insert_order_log_trigger;

CREATE TRIGGER insert_order_log_trigger
AFTER INSERT ON order_details
FOR EACH ROW
BEGIN
    -- Step 3: Extract information from the inserted row
    DECLARE new_order_details_id INT;
    DECLARE new_order_id INT;
    DECLARE new_order_date DATE;
    DECLARE new_order_time TIME;
    DECLARE new_item_id INT;
    
    # Assign values from the inserted row to variables
    SELECT NEW.order_details_id, NEW.order_id, NEW.order_date, NEW.order_time, NEW.item_id
    INTO new_order_details_id, new_order_id, new_order_date, new_order_time, new_item_id;
    
   # Insert the information into the order_log table
    INSERT INTO order_details (order_id, order_date, order_time, item_id)
    VALUES (5370, '2023-03-31', '22:15:50', 104);
END;
```
